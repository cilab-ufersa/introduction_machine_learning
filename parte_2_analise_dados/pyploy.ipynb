```python
# importando a biblioteca
import matplotlib.pyplot as plt
```

## Gráfico linha

`plt.plot`

```python
import numpy as np

plt.figure(figsize=(3,4)) # ou plt.rcParams["figure.figsize"] = (20,3)
plt.rcParams.update({'font.size': 17}) # Atualiza o tamanho da letra


x = np.array([1,2,3,4,5])
y = x**2

plt.plot(x,y, 'r--', linewidth=3, label='$x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico Linha')
plt.grid(True)
plt.legend()
plt.show()
```

## Subplots 

```python
import numpy as np

plt.rcParams["figure.figsize"] = (10,10)

fig, axes = plt.subplots(2,2)

x = np.arange(-20, 20, 1)
y = x**2 
z = 5*np.sin(x)
w = np.cos(x)
f= np.sqrt(x)


axes[0][0].plot(x,y, 'r:', linewidth=3 )
axes[0][0].set_title('Grafico 1')
axes[0][0].legend([r'$x^2$'])

axes[1][0].plot(x,z, 'b--', linewidth=3 )
axes[1][0].set_title('Grafico 2')
axes[1][0].legend([r'$5sin(x)$'])

axes[0][1].plot(x,w, 'g:', linewidth=3 )
axes[0][1].set_title('Grafico 3')
axes[0][1].legend([r'$cos(x)$'])

axes[1][1].plot(x,f, 'k-.', linewidth=3 )
axes[1][1].set_title('Grafico 4')
axes[1][1].legend([r'$\sqrt{x}$'])

plt.tight_layout()


fig = plt.gcf()
fig.savefig('figura.jpg')

plt.show()
```

## Gráfico Barra

`plt.bar`

```python

x = ['A', 'B', 'C', 'D', 'E']
y = [10, 15, 13, 17, 20]

plt.bar(x, y, color='red')

plt.xlabel('Categoria')
plt.ylabel('Valor')
plt.title('Gráfico de Barras')
plt.show()
```

## Gráfico de dispersão

`plt.scatter`

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Dispersão')
plt.show()
```

## Gráfico Histograma 

`plt.hist`

```python
import numpy as np # vamos precisar a bib numpy para gerar os valores para x

x = np.random.normal(size=1000)

plt.hist(x, bins=30)

plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.title('Gráfico de Histograma')
plt.show()
```

## Gráfico Mapa de Calor 

`plt.colorbar`

```python
import numpy as np

# Cria um conjunto de dados aleatório
data = np.random.rand(10, 10)

# Cria um heatmap a partir dos dados
heatmap = plt.pcolor(data, cmap=plt.cm.Blues)

# Adiciona uma barra de cores para o heatmap
plt.colorbar(heatmap)

# Adiciona rótulos para os eixos x e y
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Define o título do gráfico
plt.title('Heatmap')

# Exibe o gráfico
plt.show()
```

## Gráfico de Superfície 





```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Cria uma grade de valores x e y
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)

# Define uma função z(x,y) para a superfície
z = np.sin(np.sqrt(x**2 + y**2))

# Cria o gráfico de superfície
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x, y, z, cmap='coolwarm')

# Adiciona uma barra de cores para indicar os valores z
fig.colorbar(surf, shrink=0.5, aspect=5)

# Adiciona rótulos para os eixos x, y e z
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')

# Define o título do gráfico
plt.title('Surface Plot')

# Exibe o gráfico
plt.show()
```
